# Ejercicio 3

Select some points from your maps.

Create the convex hull for those points.

Turn the hull into a GDF.

Plot the hull on top of the country

```python
#I select some airports
gb_airports = infoairports[infoairports['iso_region'].str.contains('GB', na=False)]
gb_airports_filtered = gb_airports[
    (gb_airports['latitude_deg'] >= 55) & (gb_airports['latitude_deg'] <= 60 ) & (gb_airports['longitude_deg'] >= -5) & (gb_airports['longitude_deg'] <= 10 )
]

gb_airports_filtered_gdf = gpd.GeoDataFrame(
    gb_airports_filtered,
    geometry=gpd.points_from_xy(gb_airports_filtered.longitude_deg, gb_airports_filtered.latitude_deg),
    crs="EPSG:4326"  )
gb_airports_filtered_27700 = gb_airports_filtered_gdf.to_crs(england_27700.crs)

# Plot the layer
ax = england_27700.plot(color='white', edgecolor='black')
gb_airports_filtered_27700.plot(ax=ax, color='red', markersize=2)
```

```python
# Create the convex hull
hull = gb_airports_filtered_27700.unary_union.convex_hull

# Turn the hull into a GeoDataFrame
hull_gdf = gpd.GeoDataFrame(geometry=[hull], crs=gb_airports_filtered_27700.crs)

# Plot the hull on top of the country
ax = england_27700.plot(color='white', edgecolor='black')
gb_airports_filtered_27700.plot(ax=ax, color='red', markersize=2)
hull_gdf.plot(ax=ax, color='blue', alpha=0.5)
```

#Ejercico 4

Apply two spatial overlays to your maps.

```python
centroid_x_value = centroid_x.iloc[0]
UK_sub_2_below_centroid = UK_sub_2[UK_sub_2.geometry.centroid.y < centroid_y_value]
UK_sub_2_left_centroid = UK_sub_2[UK_sub_2.geometry.centroid.x < centroid_x_value]
UK_sub_2_right_centroid = UK_sub_2[UK_sub_2.geometry.centroid.x > centroid_x_value]
```

```python
ax = UK.plot(facecolor="lightgrey", edgecolor='black', linewidth=0.4, figsize=(10, 10))
UK_sub_2_below_centroid.overlay(UK_sub_2_above_centroid, how="union", keep_geom_type=False).plot(ax=ax, color="red", alpha=0.5)
UK_sub_2_right_centroid.overlay(UK_sub_2_left_centroid, how="union", keep_geom_type=False).plot(ax=ax, color="yellow", alpha=0.5)
```
